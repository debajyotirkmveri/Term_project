```python
import os
import pandas as pd

# Directory containing your CSV files
csv_directory = '/home/sysadm/Desktop/Term project/csvfile with whom i kriging'

# Iterate through all CSV files in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(csv_directory, filename)
        
        # Load the CSV file into a DataFrame
        data = pd.read_csv(file_path)
        
        # Apply the column name cleaning operations to all columns
        data.columns = data.columns.str.replace('_x', '').str.replace('_y', '').str.replace(r'\.1', '')
        
        # Save the cleaned DataFrame with the same file name
        data.to_csv(file_path, index=False)
```
