```python
fg = pd.read_csv("/home/sysadm/Downloads/2022-03-04.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame from your data
data =fg

# Pivot the data to create a grid
heatmap_data = data.pivot_table(values='No2_value(mol/m^2)', index='latitude', columns='longitude', aggfunc='first')

# Create a contour plot using plt.contourf
plt.figure(figsize=(10, 8))
contour = plt.contourf(heatmap_data.columns, heatmap_data.index, heatmap_data.values, cmap='viridis')
plt.colorbar(contour, label='NO2 Value')
plt.title('NO2 Contour Plot')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
