```python
import pandas as pd
df = pd.read_csv('/home/sysadm/Downloads/merged_data_2018_sorted_dates .csv')

def count_columns_with_null_in_intervals(dataframe):
    # Initialize counters for each interval
    interval_25_33 = 0
    interval_0_25 = 0
    interval_33_50 = 0
    interval_50_75 = 0
    interval_above_75 = 0
    
    # Iterate through all the columns except the first two
    for column_name in dataframe.columns[2:]:
        column_data = dataframe[column_name]
        
        # Calculate the total number of null values in the column
        total_null = column_data.isna().sum()
        
        # Calculate the percentage of null values in the column
        percentage_null = (total_null / len(column_data)) * 100
        
        # Check which interval the percentage falls into and update the corresponding counter
        if 25 <= percentage_null <= 33:
            interval_25_33 += 1
        elif 0 <= percentage_null < 25:
            interval_0_25 += 1
        elif 33 < percentage_null <= 50:
            interval_33_50 += 1
        elif 50 < percentage_null <= 75:
            interval_50_75 += 1
        else:
            interval_above_75 += 1
    
    return {
        '25-33%': interval_25_33,
        '0-25%': interval_0_25,
        '33-50%': interval_33_50,
        '50-75%': interval_50_75,
        'Above 75%': interval_above_75
    }


# Get the count of columns with null values in specified intervals
interval_counts = count_columns_with_null_in_intervals(df)

# Print the counts for each interval
for key, value in interval_counts.items():
    print(f"Columns with null values in {key} interval: {value}")
import matplotlib.pyplot as plt
# Create a pie plot
labels = interval_counts.keys()
sizes = interval_counts.values()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Columns with Null Values in Intervals in the year 2018')

# Display the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig("/home/sysadm/Downloads/nan_values_2018.png")
plt.show()
```
