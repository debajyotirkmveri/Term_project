# How many columns null values lies in different intervals the interval

```python
import pandas as pd

def count_columns_with_null_in_intervals(dataframe):
    # Initialize counters for each interval
    interval_25_33 = 0
    interval_0_25 = 0
    interval_33_50 = 0
    interval_50_75 = 0
    interval_above_75 = 0
    
    # Iterate through all the columns except the first two
    for column_name in dataframe.columns[2:]:
        column_data = dataframe[column_name]
        
        # Calculate the total number of null values in the column
        total_null = column_data.isna().sum()
        
        # Calculate the percentage of null values in the column
        percentage_null = (total_null / len(column_data)) * 100
        
        # Check which interval the percentage falls into and update the corresponding counter
        if 25 <= percentage_null <= 33:
            interval_25_33 += 1
        elif 0 <= percentage_null < 25:
            interval_0_25 += 1
        elif 33 < percentage_null <= 50:
            interval_33_50 += 1
        elif 50 < percentage_null <= 75:
            interval_50_75 += 1
        else:
            interval_above_75 += 1
    
    return {
        '25-33%': interval_25_33,
        '0-25%': interval_0_25,
        '33-50%': interval_33_50,
        '50-75%': interval_50_75,
        'Above 75%': interval_above_75
    }


# Get the count of columns with null values in specified intervals
interval_counts = count_columns_with_null_in_intervals(cleaned_df)

# Print the counts for each interval
for key, value in interval_counts.items():
    print(f"Columns with null values in {key} interval: {value}")
```

```python
import pandas as pd

def filter_columns_by_null_percentage(dataframe, threshold=33):
    # Initialize an empty list to store column names
    selected_columns = []

    # Iterate through all the columns except the first two
    for column_name in dataframe.columns[2:]:
        column_data = dataframe[column_name]
        
        # Calculate the total number of null values in the column
        total_null = column_data.isna().sum()
        
        # Calculate the percentage of null values in the column
        percentage_null = (total_null / len(column_data)) * 100
        
        # Check if the percentage is less than the specified threshold
        if percentage_null < threshold:
            selected_columns.append(column_name)

    # Create a new DataFrame with the selected columns
    filtered_df = dataframe[['latitude', 'longitude'] + selected_columns]
    
    return filtered_df
#cleaned_df
# Example usage:
filtered_df = filter_columns_by_null_percentage(cleaned_df, threshold=33)
filtered_df
```

```python
import pandas as pd

def create_new_dataframe_without_1350_nans(csv_file_path):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)
    
    # Calculate the number of NaN values in each column, excluding the first two columns
    total_nan_values_except_first_two = df.iloc[:, 2:].isna().sum()
    
    # Find columns with 1350 NaN values
    columns_with_1350_nans = total_nan_values_except_first_two[total_nan_values_except_first_two == 1350]
    
    # Drop columns with 1350 NaN values
    df_cleaned = df.drop(columns=columns_with_1350_nans.index)
    
    # Print the columns that were dropped
    if not columns_with_1350_nans.empty:
        print(f"Dropped columns with 1350 NaN values in {csv_file_path}:")
        print(columns_with_1350_nans)
    
    # Return the cleaned DataFrame
    return df_cleaned

# Example usage:
csv_file_path1 = '/home/sysadm/Downloads/merged_data_2021_sorted_dates.csv'
cleaned_df = create_new_dataframe_without_1350_nans(csv_file_path1)

# You can now use 'cleaned_df' for further analysis or save it to a new CSV file if needed.
```

```python
import pandas as pd

def count_columns_with_null_in_intervals(dataframe):
    # Initialize counters for each interval
    interval_25_33 = 0
    interval_0_25 = 0
    interval_33_50 = 0
    interval_50_75 = 0
    interval_above_75 = 0
    
    # Iterate through all the columns except the first two
    for column_name in dataframe.columns[2:]:
        column_data = dataframe[column_name]
        
        # Calculate the total number of null values in the column
        total_null = column_data.isna().sum()
        
        # Calculate the percentage of null values in the column
        percentage_null = (total_null / len(column_data)) * 100
        
        # Check which interval the percentage falls into and update the corresponding counter
        if 25 <= percentage_null <= 33:
            interval_25_33 += 1
        elif 0 <= percentage_null < 25:
            interval_0_25 += 1
        elif 33 < percentage_null <= 50:
            interval_33_50 += 1
        elif 50 < percentage_null <= 75:
            interval_50_75 += 1
        else:
            interval_above_75 += 1
    
    return {
        '25-33%': interval_25_33,
        '0-25%': interval_0_25,
        '33-50%': interval_33_50,
        '50-75%': interval_50_75,
        'Above 75%': interval_above_75
    }


# Get the count of columns with null values in specified intervals
interval_counts = count_columns_with_null_in_intervals(cleaned_df)

# Print the counts for each interval
for key, value in interval_counts.items():
    print(f"Columns with null values in {key} interval: {value}")
```
