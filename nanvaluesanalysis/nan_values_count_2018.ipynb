```python
import pandas as pd
from pykrige.ok import OrdinaryKriging
import numpy as np

def estimate_aod(df, date_column, nlags=20):
    # Create a DataFrame with only latitude, longitude, and the specified date column
    df1 = df[['latitude', 'longitude', date_column]]

    # Create a DataFrame with latitude and longitude where the specified date column's values are NaN
    nan_locations_df = df1[df1[date_column].isna()][['latitude', 'longitude']]

    # Create a DataFrame with latitude, longitude, and the specified date column excluding NaN values
    aod_df = df1.dropna(subset=[date_column])[['latitude', 'longitude', date_column]]

    # Prepare the data for Ordinary Kriging
    lons = np.array(aod_df['longitude'])
    lats = np.array(aod_df['latitude'])
    zdata = np.array(aod_df[date_column])

    # Perform Ordinary Kriging
    OK = OrdinaryKriging(lons, lats, zdata, variogram_model='linear', verbose=True, enable_plotting=False, nlags=nlags)
    grid_lon = np.array(nan_locations_df["longitude"])
    grid_lat = np.array(nan_locations_df["latitude"])
    z1, ss1 = OK.execute('grid', grid_lon, grid_lat)

    # Fill in NaN values in the specified date column with the interpolated values
    nan_locations_df[date_column] = z1[0]

    # Merge the two DataFrames based on their indices
    merged_df = pd.concat([aod_df, nan_locations_df], ignore_index=False)

    # Sort the DataFrame by ascending index
    merged_df = merged_df.sort_index(ascending=True)

    return merged_df

# Example usage:
# df = your_dataframe  # Replace 'your_dataframe' with your actual DataFrame
# date_column = '2018-01-02'  # Replace with the date column you want to estimate
# result_df = estimate_aod(df, date_column)

# Now 'result_df' contains the DataFrame with estimated AOD values for the specified date column.
```

```python
import pandas as pd
from pykrige.ok import OrdinaryKriging
import numpy as np

def estimate_aod(df, nlags=20):
    # Get the list of date columns with NaN values
    date_columns_with_nan = [col for col in df.columns[2:] if df[col].isna().any()]

    # Create an empty DataFrame to store the merged results
    merged_df = pd.DataFrame()

    # Iterate over the date columns with NaN values
    for date_column in date_columns_with_nan:
        # Create a DataFrame with only latitude, longitude, and the current date column
        df1 = df[['latitude', 'longitude', date_column]]

        # Create a DataFrame with latitude and longitude where the current date column's values are NaN
        nan_locations_df = df1[df1[date_column].isna()][['latitude', 'longitude']]

        # Create a DataFrame with latitude, longitude, and the current date column excluding NaN values
        aod_df = df1.dropna(subset=[date_column])[['latitude', 'longitude', date_column]]

        # Prepare the data for Ordinary Kriging
        lons = np.array(aod_df['longitude'])
        lats = np.array(aod_df['latitude'])
        zdata = np.array(aod_df[date_column])

        # Perform Ordinary Kriging
        OK = OrdinaryKriging(lons, lats, zdata, variogram_model='linear', verbose=True, enable_plotting=False, nlags=nlags)
        grid_lon = np.array(nan_locations_df["longitude"])
        grid_lat = np.array(nan_locations_df["latitude"])
        z1, ss1 = OK.execute('grid', grid_lon, grid_lat)

        # Fill in NaN values in the current date column with the interpolated values
        nan_locations_df[date_column] = z1[0]

        # Concatenate the dataframes
        aod_df = pd.concat([aod_df, nan_locations_df], ignore_index=False)

        # Sort the DataFrame by ascending index
        aod_df = aod_df.sort_index(ascending=True)

        # Concatenate the current date column's result with the merged DataFrame
        merged_df = pd.concat([merged_df, aod_df], axis=1)

    return merged_df

# Example usage:
# df = your_dataframe  # Replace 'your_dataframe' with your actual DataFrame
# result_df = estimate_aod(df)

# Now 'result_df' contains the DataFrame with estimated AOD values for all date columns with NaN values.

```
